Desafio:
Uma importadora trabalha com produtos cujos valores oscilam conforme a cotação de 3 moedas:
1. Dólar
2. Euro
3. Ouro
Analisar e atualizar diariamente os valores dos produtos (compra e venda)

Passo-a-passo:
1. Baixar webdriver do Google

2. Criar um navegador

3. Encontrar a cotação do dólar
- Abrir navegador padrão
- Clicar no campo para busca
- Buscar a cotação do dólar
- Clicar em enter
- Identificar o valor do dólar diário
- Salvar para inserir na planilha

4. Encontrar a cotação do euro

5. Encontrar a cotação do ouro

6. Importar base e atualizar dados da mesma

7. Calcular os novos dados, salvar e exportar a nova base de dados

```python
#Importando bibliotecas de interesse
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
#Criando navegador
navegador = webdriver.Chrome()
#Buscando valor do dólar
navegador.get("https://www.google.com.br/")
navegador.find_element(By.XPATH, 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input').send_keys('cotação dólar')
navegador.find_element(By.XPATH, 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input').send_keys(Keys.ENTER)
cotacao_dolar = navegador.find_element(By.XPATH, 
                       '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
#Buscando valor do euro
navegador.get("https://www.google.com.br/")
navegador.find_element(By.XPATH, 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input').send_keys('cotação euro')
navegador.find_element(By.XPATH, 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input').send_keys(Keys.ENTER)
cotacao_euro = navegador.find_element(By.XPATH, 
                       '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
#Buscando valor do ouro
navegador.get("https://www.melhorcambio.com/ouro-hoje")
cotacao_ouro = navegador.find_element(By.XPATH, 
                       '//*[@id="comercial"]').get_attribute('value')
cotacao_ouro = cotacao_ouro.replace(',', '.')

print(cotacao_dolar)
print(cotacao_euro)
print(cotacao_ouro)
```

```python
#Atualizando banco de dados
import pandas as pd

df = pd.read_excel('Produtos.xlsx')
display(df)
```

```python
#Atualizando coluna Cotação
df.loc[df['Moeda'] == 'Dólar', 'Cotação'] = float(cotacao_dolar)
df.loc[df['Moeda'] == 'Euro', 'Cotação'] = float(cotacao_euro)
df.loc[df['Moeda'] == 'Ouro', 'Cotação'] = float(cotacao_ouro)
display(df)
```

```python
#Atualizando Preços
df['Preço de Compra'] = df['Preço Original'] * df['Cotação']
df['Preço de Venda'] = df['Preço de Compra'] * df['Margem']
display(df)
```

```python
df.to_excel('Produtos Atual.xlsx', index=False)
```
